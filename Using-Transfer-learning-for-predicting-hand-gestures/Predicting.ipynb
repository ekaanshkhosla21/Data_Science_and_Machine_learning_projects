### Importing libraries

```python
from keras.models import load_model
from keras.preprocessing import image
from keras.applications.vgg16 import preprocess_input
import numpy as np
```

### Loading model and image for prediction

```python
model = load_model('handgestures.h5')
img = image.load_img('Hand gestures/Test/08_palm_moved/frame_02_08_0066.png', target_size=(224, 224))
```

### Image loaded

```python
img
```

```python
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
img_data = preprocess_input(x)
```

### Predicting 

```python
classes = model.predict(img_data)
```

```python
type(classes)
```

```python
max(classes[0])
```

```python
max_index_col = np.argmax(classes, axis=1)
```

```python
max_index_col[0]
```

```python
if(max_index_col[0]==0):
    print ('Palm')
    
if(max_index_col[0]==1):
    print ('I')
    
if(max_index_col[0]==2):
    print ('fist')
    
if(max_index_col[0]==3):
    print ('fist moved')
    
if(max_index_col[0]==4):
    print ('thumb')
    
if(max_index_col[0]==5):
    print ('index')
    
if(max_index_col[0]==6):
    print ('ok')
    
if(max_index_col[0]==7):
    print ('Palm moved')
    
if(max_index_col[0]==8):
    print ('C')
    
if(max_index_col[0]==9):
    print ('Down')
```
